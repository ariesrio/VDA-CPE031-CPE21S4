```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('/content/filtered  dataset - matplotlib.csv.csv')

# Group by Device Model and calculate average App Usage Time
avg_usage = df.groupby('Device Model')['App Usage Time (min/day)'].mean()

# Create a bar plot
plt.figure(figsize=(10, 6))
avg_usage.plot(kind='bar', color='skyblue')
plt.title('Average App Usage Time by Device Model', fontsize=16)
plt.xlabel('Device Model', fontsize=14)
plt.ylabel('Average App Usage Time (min/day)', fontsize=14)
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
# Create a box plot for Battery Drain by Operating System
plt.figure(figsize=(8, 5))
df.boxplot(column='Battery Drain (mAh/day)', by='Operating System', grid=False)
plt.title('Battery Drain Distribution by Operating System', fontsize=16)
plt.suptitle('')
plt.xlabel('Operating System', fontsize=14)
plt.ylabel('Battery Drain (mAh/day)', fontsize=14)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
import numpy as np

# Create a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['App Usage Time (min/day)'], df['Battery Drain (mAh/day)'], alpha=0.6, color='purple')
plt.title('Correlation between App Usage Time and Battery Drain', fontsize=16)
plt.xlabel('App Usage Time (min/day)', fontsize=14)
plt.ylabel('Battery Drain (mAh/day)', fontsize=14)

# Fit a trend line
z = np.polyfit(df['App Usage Time (min/day)'], df['Battery Drain (mAh/day)'], 1)
p = np.poly1d(z)
plt.plot(df['App Usage Time (min/day)'], p(df['App Usage Time (min/day)']), color='black')

plt.grid()
plt.tight_layout()
plt.show()
```
