#Question 1: Does smoking have a noticeable impact on the likelihood of having heart disease?
##The dataset contains a Smoking column and a HeartDisease column. By visualizing the relationship between smoking and heart disease, we can assess whether smokers have a higher incidence of heart disease.

```python
import matplotlib.pyplot as plt

# Question 1: Does smoking impact heart disease likelihood?
# Create a count plot of Heart Disease vs Smoking status
smoking_heart_disease = df.groupby(['Smoking', 'HeartDisease']).size().unstack().fillna(0)

# Plot the data
smoking_heart_disease.plot(kind='bar', stacked=True, color=['pink', 'skyblue'])
plt.title('Heart Disease by Smoking Status')
plt.xlabel('Smoking Status')
plt.ylabel('Number of People')
plt.xticks(rotation=0)
plt.legend(title='Heart Disease', labels=['No', 'Yes'])
plt.show()
```

*Observation: The bar chart shows that smoking is linked to a higher chance of having heart disease. More smokers have heart disease compared to non-smokers, where most people dont have heart disease. This suggests that smoking increases the risk of developing heart disease.*



# Question 2: Is there a relationship between BMI (Body Mass Index) and the general health status of individuals?

## The dataset includes a BMI column and a GenHealth column, which categorizes the general health of individuals. A scatter plot or box plot can show if individuals with higher or lower BMI tend to report poorer health.

```python
# Question 2: Is there a relationship between BMI and General Health?
# Box plot to compare BMI across different general health categories
plt.figure(figsize=(10, 6))
df.boxplot(column='BMI', by='GenHealth', grid=False, patch_artist=True,
           boxprops=dict(facecolor='skyblue', color='blue'),
           whiskerprops=dict(color='blue'), capprops=dict(color='blue'),
           medianprops=dict(color='red'))
plt.title('BMI vs General Health')
plt.suptitle('')  # Removes the default title from the figure
plt.xlabel('General Health')
plt.ylabel('BMI')
plt.xticks(rotation=45)
plt.show()
```

*The box plot shows that people with higher BMIs usually have worse general health, while those with lower BMIs tend to have better health. People in the "Poor" health group have the highest BMI, and those in the "Excellent" health group have the lowest. This suggests that a higher BMI is linked to poorer health.*
