```python
import pandas as pd
import matplotlib.pyplot as plt

#2. Read the CSV File

df = pd.read_csv('/content/Technological-Products-Sample-Data - Tech Products.csv')

df
```

#a. Read each device's date of release and sales. Show it using a line plot.

```python
plt.figure(figsize=(8, 5))

x = df['Date of Release']
y = df['Sales (USD)']

plt.plot(x, y, marker='.', color='skyblue')
plt.xlabel('Date of Release')
plt.ylabel('Sales (USD)')
plt.title("Device Sales (USD) Over Time")
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.tight_layout()
plt.show()
```

#b. Get the total sales and show line plot with the following style properties.
- Line Style dotted and Line-color should be red
- Show legend at the lower right location.
- X label name = Year
- Y label name = Sales
- Add a circle marker.
- Line marker color as red
- Line width should be 3

```python
total_sales_by_date = df.groupby('Date of Release')['Sales (USD)'].sum()

plt.figure(figsize=(10, 6))
plt.plot(total_sales_by_date.index, total_sales_by_date.values, linestyle='dotted', color='red',
         marker='o', mfc='red', linewidth=3, label='Total Sales (USD)')

plt.xlabel('Date of Release')
plt.ylabel('Sales (USD)')
plt.title('Total Device Sales on Each Release Date')
plt.legend(loc='lower right')
plt.grid(axis='y', linestyle='-', linewidth=0.5)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.show()
```

#c. Read all product sales data and show it using a multiline plot. Label them.

```python
plt.figure(figsize=(10, 6))

for model in df['Model'].unique():
    model_data = df[df['Model'] == model].groupby('Date of Release')['Sales (USD)'].sum()
    plt.plot(model_data.index, model_data.values, marker='o', label=model)

plt.xlabel('Date of Release')
plt.ylabel('Sales (USD)')
plt.title('Sales for Each Product Model by Release Date')
plt.legend(loc='upper left', fontsize=8)
plt.xticks(rotation=45, ha='right', fontsize=8)
plt.grid(True)
plt.show()
```

#d. Read each brand sales data and show it using the bar chart. Label them.

```python
sales_by_brand = df.groupby('Brand')['Sales (USD)'].sum()

plt.figure(figsize=(10, 6))
sales_by_brand.plot(kind='bar', color='skyblue', alpha=0.7)

plt.xlabel('Brand')
plt.ylabel('Total Sales (USD)')
plt.title('Total Sales by Brand')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#e. Read each device and show it using the histogram to see the most common Country of Origin. Label them.

```python
plt.figure(figsize=(8, 5))
df['Country of Origin'].value_counts().plot(kind='bar', color='skyblue', alpha=0.8)

plt.xlabel('Country of Origin')
plt.ylabel('Device Count')
plt.title('Distribution of Device Origins')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#f. Create a pie chart that shows each brand Country of Origin. Label them.

```python
brands_by_country = df.groupby('Country of Origin')['Brand'].nunique()

plt.figure(figsize=(8, 8))
brands_by_country.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Pastel1.colors)

plt.ylabel('')
plt.title('Brand Distribution by Country of Origin')
plt.show()
```
